```python
#3. Plot subplots for categorical features in titanic dataset , consisting of rows = 3 and columns = 2 format.

titanicdata = pd.read_csv(r'C:\Users\Insta\Desktop\End2End\endtoenddatasets\titanic.csv')
```

```python
titanicdata.head()
```

```python
viz = plt.figure(figsize=(30,10))
x = 1
for titanic in titanicdata.select_dtypes("object"):
    plt.subplot(3,2,x)
    titanicdata[titanic].value_counts().plot.bar(color = 'magenta')
    x = x + 1
plt.tight_layout()
plt.show()
    
```

```python
#4. In the diabetes dataset, visualize the numerical variables using histogram and make use of subplots while plotting.

diabetesdata = pd.read_csv(r'C:\Users\Insta\Desktop\End2End\endtoenddatasets\diabetes.csv')
```

```python
diabetesdata.head()
```

```python
viz = plt.figure(figsize=(10,10))
i = 1
for diabetes in diabetesdata.select_dtypes(["int64","float64"]):
    plt.subplot(3,3,i)
    plt.hist(diabetesdata[diabetes])
    i = i + 1
plt.tight_layout()
plt.show()
```

```python
#9. In Boston housing dataset, plot scatter of 'LSTAT' Vs 'MEDV', 'INDUS' Vs 'MEDV' and 'TAX' and 'MEDV'.

bostondata = pd.read_csv(r'C:\Users\Insta\Desktop\End2End\endtoenddatasets\boston_house_price.csv')
```

```python
plt.scatter(bostondata['LSTAT'],bostondata['MEDV'], color='green')

plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.title("LSTAT Vs MEDV PLOT: \n",fontsize=10)
plt.show()
```
